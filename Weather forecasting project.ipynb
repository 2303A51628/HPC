```python
import time
import random
import multiprocessing
import matplotlib.pyplot as plt

def predict_temperature(city):
    temp = random.uniform(20, 40)

    # Heavy computation simulation
    for _ in range(10_000_000):
        temp += 0.000001

    return (city, round(temp, 2))

if __name__ == "__main__":
    cities = [f"City_{i}" for i in range(1, 21)]

    start_time = time.perf_counter()

    with multiprocessing.Pool(processes=4) as pool:
        results = pool.map(predict_temperature, cities)

    end_time = time.perf_counter()

    # Separate data
    city_names = [r[0] for r in results]
    temps = [r[1] for r in results]

    # Print results
    for city, temp in results:
        print(f"{city}: Predicted Temp = {temp}°C")

    print("\nParallel Execution Time:", round(end_time - start_time, 2), "seconds")


    plt.figure()
    plt.bar(city_names, temps)

    plt.xlabel("Cities")
    plt.ylabel("Temperature (°C)")
    plt.title("Predicted Temperature for Cities")

    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

# Function to simulate heavy weather calculation
def predict_temperature(city):
    temp = random.uniform(20, 40)

    # Heavy computation simulation
    for _ in range(10_000_000):
        temp += 0.000001

    return (city, round(temp, 2))

cities = [f"City_{i}" for i in range(1, 21)]

start_time = time.perf_counter()

results = []
for city in cities:
    results.append(predict_temperature(city))

end_time = time.perf_counter()

# Print results
for city, temp in results:
    print(f"{city}: Predicted Temp = {temp}°C")

print("\nSequential Execution Time:", round(end_time - start_time, 2), "seconds")


city_names = [r[0] for r in results]
temps = [r[1] for r in results]

plt.figure()
plt.bar(city_names, temps)

plt.xlabel("Cities")
plt.ylabel("Temperature (°C)")
plt.title("Sequential Temperature Prediction")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
